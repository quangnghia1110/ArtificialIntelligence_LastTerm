```python
class CSP_Problem():
    def __init__(self):
        self.variables = []
        self.domain = {}
        self.neighbours = {}
        self.constraints = {}
        
    def get_state(self): pass

    def action(self): pass
    
```

```python
class Kakuro_Problem(CSP_Problem):
    def __init__(self, number_of_rows, number_of_columns, horizontal, vertical):
        CSP_Problem.__init__(self)
        self.number_of_rows = number_of_rows
        self.number_of_columns = number_of_columns
        self.horizontal = horizontal
        self.vertical = vertical
        self.type_cell = {}
        for i in range(number_of_rows):
            for j in range(number_of_columns):
                self.type_cell[(i, j)] = 'blank_cell'
        #For the horizontal values
        for i in range(self.number_of_rows):
            for j in range(self.number_of_columns):
                #Find sum clues
                if self.horizontal[i][j] != "#" and self.horizontal[i][j] != 0:
                    u_variable_name = "UR" + str(i) + "," + str(j)
                    self.type_cell[(i, j)] = 'constraint_cell'
                    #Add the variable
                    self.variables.append(u_variable_name)
                    self.neighbours[u_variable_name] = []
                    k = j + 1
                    ct = 0
                    #set the domain, constraints, neighbours and variables of the block
                    while k < number_of_columns:
                        if self.horizontal[i][k] == 0:
                            x_variable_name = "X" + str(i) + "," + str(k)
                            self.type_cell[(i, k)] = 'value_cell'
                            self.variables.append(x_variable_name)
                            self.neighbours[x_variable_name] = []
                            self.neighbours[x_variable_name].append(u_variable_name)
                            self.neighbours[u_variable_name].append(x_variable_name)
                            self.domain[x_variable_name] = [l for l in range(1,min(int(self.horizontal[i][j]),10))]
                            self.constraints[u_variable_name+x_variable_name] = ct
                            ct += 1
                        else:
                            break
                        k += 1
                    self.domain[u_variable_name] = [lst for lst in permutations([1,2,3,4,5,6,7,8,9],k-j-1) if sum(lst) == int(self.horizontal[i][j])]
                    j = k # Không có ý nghĩa vì giá trị của j sẽ tự động cập nhật lại theo dãy range khi thực hiện lần lặp tiếp theo
                    
        #For the vertical values
        for j in range(self.number_of_columns):
            for i in range(self.number_of_rows):
                #find the sum clues
                if self.vertical[i][j] != "#" and self.vertical[i][j] != 0:
                    u_variable_name = "UD"+str(i)+","+str(j)
                    self.type_cell[(i, j)] = 'constraint_cell'
                    #Add the variable
                    self.variables.append(u_variable_name)
                    self.neighbours[u_variable_name] = []
                    k = i + 1
                    ct = 0
                    #set the domains constraints neighbours variables in block
                    while k < number_of_rows:
                        if self.vertical[k][j] == 0:
                            x_variable_name = "X"+str(k)+","+str(j)
                            if x_variable_name not in self.variables:
                                self.variables.append(x_variable_name)
                            if x_variable_name in self.neighbours:
                                self.neighbours[x_variable_name].append(u_variable_name)
                            else:
                                self.neighbours[x_variable_name] = [u_variable_name]
                            self.neighbours[u_variable_name].append(x_variable_name)
                            if x_variable_name in self.domain:
                                self.domain[x_variable_name] = list(set(self.domain[x_variable_name]).intersection(set([l for l in range(1,min(int(self.vertical[i][j]),10))])))
                            else:
                                self.domain[x_variable_name] = [l for l in range(1,min(int(self.vertical[i][j]),10))]
                            self.constraints[u_variable_name+x_variable_name] = ct
                            ct += 1
                        else:
                            break
                        k += 1
                    self.domain[u_variable_name] = [lst for lst in permutations([1,2,3,4,5,6,7,8,9],k-i-1) if sum(lst) == int(self.vertical[i][j])] 
                    i = k
                    
    def order_domain_values(self, variable):
        return self.domain[variable]
    
    #if assigning the value to variable is consistent with previous assignments
    #returns True
    #else return False    
    def consistent(self, variable, value, assignment):
    #checking if the variable is a empty cell variable or the one assigned with constraints
        if variable[0] == "U":
            for neighbour in self.neighbours[variable]:
                if neighbour in assignment:
                    if assignment[neighbour] != value[self.constraints[variable+neighbour]]:
                        return False
            return True
        else:
            for neighbour in self.neighbours[variable]:
                if neighbour in assignment:
                    if assignment[neighbour][self.constraints[neighbour+variable]] != value:
                        return False
            return True
            
    def is_satisfy_constraint(self, xi, valueXi, xj, valueXj):
        satisfaction = False
        if xj[0] == "U": # Biến ràng buộc
            if valueXi == valueXj[self.constraints[xj+xi]]:
                satisfaction = True
        else: # Biến giá trị
            if valueXi[self.constraints[xi+xj]] == valueXj:
                satisfaction = True
        return satisfaction
        
    #finds a variable which is not assigned and returns it
    def select_unassigned_variable(self, assignment):
        for variable in self.variables:
            if variable not in assignment:
                return variable
      
    def get_state(self, assignment):
        CSP_Problem.get_state(self)
        return tuple(assignment)
        
    def fill_in_cell(self, assignment): #### Cần làm lại dựa trên assignment
        CSP_Problem.action(self)
        for variable in assignment:
            i, j = int(variable[-3]), int(variable[-1])
            if self.type_cell[(i, j)] == 'value_cell':
                self.horizontal[i][j] = assignment[variable]
                self.vertical[i][j] = assignment[variable] 
    
    def show(self):
        print("____Horizontal_______")
        print(self.horizontal)
        print("____Vertical_______")
        print(self.vertical)
        # Cần show ra cái gì
    
    def show_img(self, figsize):
        # Vẽ lưới Sudoku
        fig = plt.figure(figsize=figsize)
        ax = fig.add_subplot(111)

        # Tạo lưới trên đó
        ax.set_xticks(np.arange(-0.5, self.number_of_columns, 1))
        ax.set_yticks(np.arange(-0.5, self.number_of_rows, 1))
        ax.set_xlim([-0.5, self.number_of_columns-0.5])
        ax.set_ylim([-0.5, self.number_of_rows-0.5])
        ax.invert_yaxis()
        ax.grid(which='both', color='black', linestyle='-', linewidth=2)
        
        for i in range(self.number_of_rows):
            for j in range(self.number_of_columns):
                if self.type_cell[(i, j)] == 'value_cell': 
                    if self.horizontal[i][j] != 0:
                        ax.text(j, i, self.horizontal[i][j], fontsize=30,color = 'red', ha='center', va='center')
                else:
                    ax.add_patch(plt.Rectangle((j-0.5, i-0.5), 1, 1, linewidth=2, edgecolor='black', facecolor='gray'))
                    ax.plot([j-0.5, j+0.5], [i-0.5, i+0.5], color='black', linestyle='-', linewidth=2)
                    if self.type_cell[(i, j)] == 'constraint_cell':
                        if self.horizontal[i][j] != '#':
                            ax.text(j, i, self.horizontal[i][j], fontsize=20, ha='left', va='bottom')
                        if self.vertical[i][j] != '#':
                            ax.text(j, i, self.vertical[i][j], fontsize=20, ha='right', va='top')
        # Đặt tiêu đề và hiển thị
        ax.set_title('Kakuro Game Board')
        ax.set_aspect('equal')
        plt.show()
```

```python
kakuro_game = Kakuro_Problem(rows, columns, Horizontal, Vertical)
```

```python
class Problem_solver:
    def __init__(self): pass
    def train(self, problem: Type[CSP_Problem]): pass
    def solve(self): pass
```

```python
class backtracking_search_with_MAC_solver(Problem_solver):
    def __init__(self):
        Problem_solver.__init__(self)
        
    def train(self, problem: Type[CSP_Problem]):
        Problem_solver.train(self, problem)
        self.csp_problem = problem
    
    def solve(self):
        Problem_solver.solve(self)
        self.AC_3(self.csp_problem)
        return self.backtracking_search_with_MAC(self.csp_problem)
    
    #The AC-3 algorithm returns false if the domain becomes empty of some variable
    def AC_3(self, csp, queue=None):
        if queue == None:
            queue = [(xi,xj) for xi in csp.variables for xj in csp.neighbours[xi]]
        while queue:
            xi,xj = queue.pop(0)
            if self.revise(csp, xi, xj):
                if len(csp.domain[xi]) == 0:
                    return False
                #add the neighbour arcs - {xj} in queue if some value is removed
                for neighbour in csp.neighbours[xi]:
                    if neighbour != xj: # Debug: ???? # Có ảnh hưởng đến MAC hay không
                        queue.append((neighbour, xi))
        return True
    ### Đưa hàm remove_inconsistency_value từ class problem -> solver ?????
    def revise(self, csp, xi, xj):
        revised = False
        for valueXi in csp.domain[xi]:
            remove = True
            for valueXj in csp.domain[xj]:
                if csp.is_satisfy_constraint(xi, valueXi, xj, valueXj):
                    remove = False
                    break
            if remove:
                csp.domain[xi].remove(valueXi)
                revised = True
        return revised   

    # Maintaining arc consistency by calling AC-3 with the queue with only arcs of
    #neighbours of variable
    def MAC(self, csp, variable, assignment):
        queue = [(neighbour,variable) for neighbour in csp.neighbours[variable] if neighbour not in assignment]
        return self.AC_3(csp, queue)
    
    #recursive backtracking with mac which returns solution or None
    def recursive_backtracking_with_MAC(self, csp,assignment):
        #global backtracking_count
        #global consistency_checks_by_backtracking
        #backtracking_count += 1
        #If all values assigned then return the assignment
        if len(assignment) == len(csp.variables):
            return assignment
        variable = csp.select_unassigned_variable(assignment)
        for value in csp.order_domain_values(variable):
            #consistency_checks_by_backtracking += 1
            if csp.consistent(variable, value, assignment):
                #assign a value to variable
                assignment[variable] = value
                #set the domain of variable as the assigned value
                csp.domain[variable] = [value]
                #save the domain
                saved_domain = deepcopy(csp.domain)
                #prune the domain and if no domain is empty
                if self.MAC(csp, variable, assignment):
                    #recursively call backtracking
                    result = self.recursive_backtracking_with_MAC(csp,assignment)
                    if result != None:
                        return result
                #restore domain back 
                csp.domain = deepcopy(saved_domain)
        #restore the assignment
        if variable in assignment:
            del assignment[variable]
        return None
    
    def backtracking_search_with_MAC(self, csp):
        return self.recursive_backtracking_with_MAC(csp,{})
```

```python
solver = backtracking_search_with_MAC_solver()
solver.train(kakuro_game)
assignment = solver.solve()
print(assignment)
kakuro_game.fill_in_cell(assignment)
kakuro_game.show_img((10, 10))
```
